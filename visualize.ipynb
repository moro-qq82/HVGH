```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dance000.txt', delim_whitespace=True, header=None)
df
```

```python
plt.plot(df[0])
```

```python
plt.figure(figsize=(16, 6))
for col in df.columns:
    plt.plot(df.index, df[col], label=f'col {col}')
plt.xlabel('行番号')
plt.ylabel('値')
plt.title('全列の可視化（凡例付き）')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
```
